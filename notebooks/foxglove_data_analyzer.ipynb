```python
gps_messages = [
    (message.latitude, message.longitude)
    for topic, record, message in client.get_messages(
        device_id=device_id,
        start=coverage[0]["start"],
        end=coverage[0]["end"],
        topics=["/navsat_topic"],
    )
]
pd.DataFrame(gps_messages, columns=["lat", "lon"]).head()
```

```python
import folium

figure = folium.Figure(width=640, height=480)
map = folium.Map(location=gps_messages[0], zoom_start=200, width="100%")

# Split GPS messages into chunks of 1000 messages
gps_messages_1 = gps_messages[:300]
gps_messages_2 = gps_messages[300:600]
gps_messages_3 = gps_messages[600:900]


# Add the first chunk of GPS messages to the map and color them red

folium.PolyLine(
    locations=gps_messages_3,
    weight=10,
    color="red",
).add_to(map)
map.add_to(figure)

# Add the second chunk of GPS messages to the map and color them blue

folium.PolyLine(
    locations=gps_messages_2,
    weight=10,
    color="blue",
).add_to(map)
map.add_to(figure)

# Add the third chunk of GPS messages to the map and color them green

folium.PolyLine(
    locations=gps_messages_1,
    weight=10,
    color="green",
).add_to(map)
map.add_to(figure)
```
